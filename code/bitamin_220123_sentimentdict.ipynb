```python
# knusl.py 파일 수정 
import json
def data_list(wordname):	
  with open('/content/KnuSentiLex/data/SentiWord_info.json', encoding='utf-8-sig', mode='r') as f:
    data = json.load(f)
  result = ['None','None']	
  for i in range(0, len(data)):
    if data[i]['word'] == wordname:
      result.pop()
      result.pop()
      #result.append(data[i]['word_root']) #원래는 어근인데 원래 단어가 더 나을 것 같아서 수정
      result.append(data[i]['word'])
      result.append(data[i]['polarity'])	
  
  r_word = result[0]
  s_word = result[1]
            
  #print('어근 : ' + r_word)
  #print('극성 : ' + s_word)		
  
  return r_word, s_word
```

```python
# 리뷰 형태소에 대해서 다시 
import pandas as pd
df = pd.read_csv('/content/drive/Shareddrives/컨퍼런스/김다희/topics_console_0119.csv', encoding='utf-8', index_col=0)

word_list = []
for i in range(len(df)):
  temp_list = df.iloc[i, 23:].tolist() 
  temp_list = [x for x in temp_list if pd.isnull(x)==False] # nan 제거 
  word_list.append(temp_list)
```

```python
# 리뷰별 감성사전 돌리기(모든 단어)
# 우리가 할 때는 리뷰별로 토픽별 감성어휘를 먼저 선정하고 (by 민하님 노션정리본) 그 단어들에 대해서만 돌리기  
# 일단은 KNU 베이스, 가구별로 

wordslist = word_list[:100] # 일단은 리뷰 100개에 대해서만 

res = []
for words in wordslist: 
  tmp = []
  for word in words:
    word = word.strip(" ")
    tmp.append(data_list(word))
  res.append(tmp)
```

### __3. KAKAO PORORO__
* https://www.dinolabs.ai/396
* https://kakaobrain.github.io/pororo/text_cls/review_score.html
* 뽀로로(Pororo): 카카오 자연어처리 오픈소스 라이브러리
* 리뷰점수화(Review Scoring): 평점이 없는 리뷰데이터 점수화 가능(1.0~5.0)
* 단, 리뷰 길이 제한 있음 -> 512개(토큰 수) 
* 우리는 이걸 토픽별 감성점수를 산출하는데 사용
* (1) 리뷰별로 토픽키워드별 문장 분류 
* (2) 리뷰별로 토픽키워드별 감성점수 산출


```python
# 뽀로로 설치
!pip install pororo
from pororo import Pororo

# 'review'를 parameter로 입력하여 클래스 불러오기
review = Pororo(task='review', lang='ko')
review
```

```python
# 기본 전처리(ex. 별점5.0점 -> 5.0)
def preprocess_df(df):
  
    #basic
    df = df.iloc[:, 1:]
    df = df.dropna(subset=['user'])
    df = df.dropna(subset=['review'])
    df = df.drop_duplicates()
    df = df.reset_index(drop=True)

    #options & help_count
    df["options"] = df["options"].fillna("")
    help = df[df.options.str.contains("명에게 도움이 되었습니다")]["options"].str.extract("(\d+)", expand = False)
    df.loc[help.index, "help_count"] = help
    df.loc[help.index, "options"] = ""
    help = df[df.options.str.contains("도움이 돼요")].index
    df.loc[help, "options"] = np.NaN
    help = df[df["options"] == ""].index
    df.loc[help, "options"] = np.NaN
    
    #scores
    df.total_score = df.total_score.str.extract("(\d\.\d)", expand = False).apply(lambda x: float(x))
    df.dur_score = df.dur_score.str.extract("(\d\.\d)", expand = False).apply(lambda x: float(x))
    df.price_score = df.price_score.str.extract("(\d\.\d)", expand = False).apply(lambda x: float(x))
    df.design_score = df.design_score.str.extract("(\d\.\d)", expand = False).apply(lambda x: float(x))
    df.delivery_score = df.delivery_score.str.extract("(\d\.\d)", expand = False).apply(lambda x: float(x))
    df.item_score = df.item_score.str.extract("(\d\.\d)", expand = False).apply(lambda x: float(x))

    #item_sale_price
    df["item_sale_price"] = df["item_sale_price"].apply(lambda x: int(x.split(",")[0] + x.split(",")[1]))
  
    return df

import pandas as pd
import numpy as np 
df = pd.read_csv('/content/drive/Shareddrives/컨퍼런스/김다희/topics_console_0119.csv', encoding='utf-8')
df = df[:100] # 일단 일부만
df = preprocess_df(df)
df[['review', 'total_score']]
```

```python
# 실제 평점 vs 뽀로로 평점 
for i in range(len(df)):
  r = df['total_score'][i]
  r_po = review(df['review'][i])
  print('실제 평점:', r, '뽀로로 평점:', r_po)

print('MSE:', (r-r_po)**2/len(df))
```

```python
# 실제 평점 vs 무지성 5.0점 
for i in range(len(df)):
  r = df['total_score'][i]
  r2 = 5
  print('실제 평점:', r, '무지성:', 5)

print('MSE:', (r-r2)**2/len(df))
```
